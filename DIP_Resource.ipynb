```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read the image and get it in matrix form for futher manipulation using imread function

img_raw = cv2.imread("/content/brainmrig.jpg")

```

```python
#The first thing that must have come to your mind will why converting image from BGR to RGB. 
img = cv2.cvtColor(img_raw, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
#Image negative we use outputImage = cv2.bitwise_not(source_image)

Imagem = cv2.bitwise_not(img)
plt.imshow(Imagem)
```

```python
#Log transformation function

#Output_image =  log(1+input_image) where c is a constant mostly 1

s_img = cv2.imread("/content/brainmrig.jpg")

img = cv2.cvtColor(s_img, cv2.COLOR_BGR2RGB)

plt.imshow(s_img)

```

```python
#lets apply the log transformation function to see the change in the image.

img_1 = np.uint8(np.log1p(s_img))


```

```python
#In this line of code, np.uint8 is used to convert all the array elements into unsingned 8-bit integer.
#Then np.log1p is used for log transformation 1p means that (1 + source_img)

thresh = 1
img_2 = cv2.threshold(img_1,thresh,255,cv2.THRESH_BINARY)[1]


```

```python
#This function is used to convert the imgage into a binary image.

plt.imshow(img_2)
```

```python
#Power law or gamma transformation

#output_img = c(input_img)^
 #where gamma = 1 and value of gamma between 0 and 1 .

gamma = 1
img_2 = np.power(img,gamma)
plt.imshow(img_2,cmap='gray')


```

```python
gamma = 0.2
img_2 = np.power(img,gamma)
plt.imshow(img_2,cmap='gray')


```
